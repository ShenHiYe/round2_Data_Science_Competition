```python
import pandas as pd

df_wyckoff = pd.read_csv("wyckoff_selected.csv")
df_svm = pd.read_csv("SVM_selected.csv")

# --- Lấy danh sách mã trong SVM ---
svm_tickers = set(df_svm["ticker"].unique())

# --- Tạo cột SVM_boosted, nếu mã có trong SVM thì được 3 điểm, không có thì 0 điểm ---
df_wyckoff["SVM_boosted"] = df_wyckoff["ticker"].apply(
    lambda x: 3 if x in svm_tickers else 0
)

# --- Cập nhật score, nếu mã có xuất hiện trong SVM thì cộng 3 điểm, không thì không cộng điểm nào ---
df_wyckoff["score"] = df_wyckoff["score"] + df_wyckoff["SVM_boosted"]

# --- Sắp xếp theo score giảm dần ---
df_final = df_wyckoff.sort_values("score", ascending=False)
df_final.to_csv("final.csv", index=False, encoding="utf-8-sig")

print("Đã tạo final.csv với cột SVM_boosted và đã sắp xếp theo score")
print(df_final.head(20))
```
